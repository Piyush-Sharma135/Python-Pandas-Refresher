# Numeric Series Aggregation


```python
import pandas as pd
import numpy as np 
transactions = pd.read_csv("/home/picassa240/Maven Pandas Numpy/Pandas Course Resources/retail/transactions.csv")
transactions_series = pd.Series(transactions["transactions"])

transactions_series.head()
transactions_series.iloc[0:5]
```

```python
transactions_series.count()
```

```python
transactions_series.quantile(0.5)
```

```python
# Note that for a small dataset we need not always have an exact quantile value. For eg
print(transactions_series.iloc[:5].quantile(0.4))
# Note that 2035 doesnot exist in selected dataset 

# To get nearest quantile value
print(transactions_series.iloc[:5].quantile(0.4, interpolation ='nearest'))
# 2111 is actually our 50th percentile. In large datasets it wont be an issue. 
```
