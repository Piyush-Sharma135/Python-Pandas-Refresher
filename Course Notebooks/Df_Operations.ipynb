# Drop Rows with Missing Values/ NULL/ NaN

```python
# This removes if there is NaN in any row
nba.dropna()
# By default the parameters are set to 'any' i.e nba.dropna(how = "any")

# To drop rows where ALL the values are missing 
nba.dropna(how = "all")
```

```python
# SUBSET parameter removes all the rows where there is a NaN/Null in the specified column
nba.dropna(subset=["College"])
# Row removed where College value is NaN
nba.dropna(subset=["College", "Salary"])
```

# Fill Missing values

# Converting datatypes using astype() Method
1. Column being converted MUST NOT have MISSING values else error

```python
nba = pd.read_csv("~/Pandas/Data Set/nba.csv").dropna(how = "all")
nba.tail()
```

```python
nba["Age"].hasnans
```

```python
# Age is float
nba["Age"].astype("int")
nba["Age"].astype(int)
# NOTE : astype method doesn't take inplace as a parameter so explicit assignment is better way to make changes to DF
nba["Age"] = nba["Age"].astype("int")
nba
```

```python
nba["Salary"].hasnans
```

```python
# nba["Salary"].astype("int") gives error due to NaN
nba["Salary"] = nba["Salary"].fillna(0)
nba["Salary"] = nba["Salary"].astype(int)
nba.head()
```

# Assigning RANK to Values using rank() Method

```python
nba = pd.read_csv("~/Pandas/Data Set/nba.csv").dropna(how = "all")
nba.head()
```

```python
# Replacing NaN with 0 so that ranking is achieved properly
nba["Salary"] = nba["Salary"].fillna(0).astype(int)

# Rank values by salaries in desc order
nba["Salary_Rank"] = nba["Salary"].rank(ascending=False).astype(int)
```
