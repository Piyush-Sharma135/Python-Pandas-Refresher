```python
bigmac = pd.read_csv("~/Pandas/Data Set/bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace =True)
bigmac.head(3)
```

```python
# Pass a tuple for row 
bigmac.loc[("Price in US Dollars",)]
```

```python
# To access multiple cols]
bigmac.loc[("Price in US Dollars",), ("2010-01-01","Sri Lanka"):("2011-02-01", "Ukraine")]
```

# PIVOT Method
- taking values in col and using them as column header

```python
import pandas as pd
sales = pd.read_csv("~/Pandas/Data Set/salesmen.csv",parse_dates=["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
print(len(sales))
print(sales["Salesman"].value_counts())
print(sales.head())

# we have 5 unique values in salesman column
```

```python
print(len(sales.pivot(index="Date", columns="Salesman", values = "Revenue")))
sales.pivot(index="Date", columns="Salesman", values = "Revenue")
```

# pivot_table( ) Method




```python
foods = pd.read_csv("~/Pandas/Data Set/foods.csv")
foods.head(3)
```

```python
foods.pivot_table(
    values="Spend",
    index="City",
    aggfunc="sum", 
    margins= True)
    
```

```python
# Filtering the df using query method before applying PT method

(foods.query('Item in ["Burger", "Sushi"]')
.pivot_table(
    values="Spend",
    index="City",
    aggfunc="mean", 
    margins= True))
```

```python
foods.pivot_table(values="Spend",index="Item",aggfunc="sum")
```

```python
# MultiIndex and Multi Aggregation in Pivot Table

foods.pivot_table(
    values="Spend",
    index=["Gender", "Item"],
    aggfunc=('min', 'max')
    )

    # Can also pass dictionary in 
    # aggfunc = ( 
    # {"col1" : ["sum", "mean"], 
    # "col2" : "max" } )
```

```python
foods.pivot_table(values="Spend",index=["Gender","Item"],columns=["Frequency","City"],aggfunc="sum")
```

```python
foods.pivot_table(values="Spend",index=["Gender","Item"],columns="City",aggfunc="min").head(3)
```

```python
# Calling PT method directly on pd . But then we need to provide parameter data with some value
pd.pivot_table(data = foods,values="Spend",index=["Gender","Item"],columns="City",aggfunc="min").head(10)
```

# Pivot Table Heatmaps



```python
(pd
.pivot_table(
    data = foods,
    values="Spend",
    index=["Gender","Item"],
    columns="City",
    aggfunc="min")
.head(10)
.style.background_gradient(cmap = "RdYlGn", axis = 0))
```
