# Rename Index Lables/ Columns
Using mapper and index parameter

```python
bond = pd.read_csv("~/Pandas/Data Set/jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# Renaming Index Labels
# Passing a Python Dictionary to mapper
bond.rename(mapper = {
    "GoldenEye":"Golden Eye", 
    "The World Is Not Enough": "Best Bond Movie Ever"
    })
```

```python
# By default Pandas looks for Dictionary Keys in Index Labels , but we can also specify explicitly using axis parameter
bond.rename(mapper = {
    "GoldenEye":"Golden Eye", 
    "The World Is Not Enough": "Best Bond Movie Ever"
    }, axis =0)

bond.rename(mapper = {
    "GoldenEye":"Golden Eye", 
    "The World Is Not Enough": "Best Bond Movie Ever"
    }, axis="rows", inplace=True)
```

```python
bond
```

```python
# RENAMING COLUMN NAMES

bond.rename(mapper = {
   "Year" : "Release Year",
   "Box Office" : "Revenue" 
}, axis =1)

bond.rename(mapper = {
   "Year" : "Release Year",
   "Box Office" : "Revenue" 
}, axis ="columns")

bond.rename(columns = {
   "Year" : "Release Year",
   "Box Office" : "Revenue" 
}, inplace = True)

bond.head(0)
```

```python
# Another method to RENAME COLUMNS
# Pass a LIST with new col names BUT
# Length of List MUST BE = No of COLUMNS
bond.columns = ["Year of Release", "Actor", "Director", "Revenue Generated", "Budget", "Lead Actor Salary"]
bond.head(0)

# This is used when renaming all columns
```

# Using apply( ) method to apply function on column values

```python
bond = pd.read_csv("~/Pandas/Data Set/jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def convert_string_to_millions(number):
    return str(number) + "MILLIONS"

columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_string_to_millions) 

bond.head() 
```

# Applying fn on row values (Horizontally) and generating a custom column

```python
bond = pd.read_csv("~/Pandas/Data Set/jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)


def good_movie(row):
    actor = row[1]
    budget = row[4]
    if actor == "Pierce Brosnan":
        return "The Best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"


bond.apply(good_movie,axis = "columns")
# By setting axis = col we pass the entire row values as a list to the function
```
