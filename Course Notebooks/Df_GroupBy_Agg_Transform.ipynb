# Grouping By Multiple Cols

- By default, every column a DataFrame that is grouped by becomes a layer in a Multi-Index.

```python
fortune = pd.read_csv("~/Pandas/Data Set/fortune1000.csv", index_col="Rank")
sectors= fortune.groupby(["Sector","Industry"]) # set as_index = False to prevent grouped columns from becoming indices
fortune.head(3)
```

```python
sectors= fortune.groupby(["Sector","Industry"], as_index= False)
sectors.size()
```

```python
sectors[["Profits", "Revenue"]].sum()

# Another way to get same result

fortune.groupby(["Sector","Industry"], as_index=False)[["Profits", "Revenue"]].sum()

# We can also use the agg function  
```

```python
# Filter above result for Wholesale sector

fortune.groupby(["Sector","Industry"], as_index=False)[["Profits", "Revenue"]].sum().query("Sector == 'Wholesalers'")
```

```python
# Same output using agg method

(fortune
.groupby(["Sector","Industry"], as_index=False)
.agg({
    "Profits" : "sum", 
    "Revenue" : "sum"})
.query("Sector == 'Wholesalers'"))
```

```python
# Better to apply filter first and reduce the dataset size before applying mathematical operations

(fortune
.query("Sector == 'Wholesalers'")
.groupby(["Sector","Industry"], as_index=False)
.agg({
    "Profits" : "sum", 
    "Revenue" : "sum"})
)
```

```python
sectors[["Revenue", "Profits"]].max()
```

# Aggregation using  .agg( ) Method

- To perform multiple aggregations on a “groupby” object

```python
fortune = pd.read_csv("~/Pandas/Data Set/fortune1000.csv", index_col="Rank")
sectors= fortune.groupby("Sector")
fortune.head(3)
```

```python
# Takes input as a dictionary
#  Key : Col_name and Value : Operation to perform

sectors.agg({
    "Revenue": ["sum", "mean"],
    "Profits"  : "sum",
    "Employees" : "mean"})
```

```python
# Creating named aggregation columns and sorting by revenue

(fortune
.groupby("Sector")
.agg(
    Revenue_Sum = ("Revenue","sum"),
    Profit_Sum = ("Profits", "sum"),
    Employee_Mean = ("Employees","mean"))
.sort_values(by="Revenue_Sum", ascending=False))
```

```python
# Performing multiple operations on multiple columns

sectors.agg(["size", "sum", "mean"])
```

# Transform method

- .transform() allows us to create a column storing aggregations by groups while keeping rows intact
- used when we wish to compare each day's sales with mean sales
- we basically calculate group level statistic and perform row level analysis

```python
fortune = pd.read_csv("~/Pandas/Data Set/fortune1000.csv", index_col="Rank")
# sectors= fortune.groupby("Sector")
fortune.head()
```

```python
# Get a list of bottom 5 performing companies by their revenue

(fortune.assign(
    avg_revenue = fortune.groupby(["Sector"])["Revenue"]
                .transform("mean").round(2),

revenue_difference = lambda x : x ["Revenue"] - x["avg_revenue"]

)).sort_values(by="revenue_difference").head()
```

# Iterating through Groups

```python
fortune = pd.read_csv("~/Pandas/Data Set/fortune1000.csv", index_col="Rank")
sectors= fortune.groupby("Sector")
fortune.head(3)
```

```python
# Find companies with max profit in each sector
# sectors["Profits"].max() gives sector wise max profit but we dont get full row level detail
# we create an empty df with col headers same as fortune df
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
# to iterate over a groupby object, we need 2 variables
# 1st var to represnt the group
# 2nd var to represent the df
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1,"Revenue") # get max value based on col "Revenue"
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

```python
# City level analysis
cities = fortune.groupby("Location")
df = pd.DataFrame (columns= fortune.columns)

for city, data in cities:
    highest_revenue_in_city = data.nlargest(1,"Revenue")
    df = df.append(highest_revenue_in_city)

df
```
